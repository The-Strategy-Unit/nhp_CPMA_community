```sql
--- Selecting relevant columns from apc table and adding in flags for first batch of mitigator (split mitigators into 2 sets as was running out of memory)

SET spark.sql.shuffle.partitions = auto;

CREATE OR REPLACE TABLE strategyunit.default.SLtemp_apc_mitigator_table AS

SELECT
a.epikey,
a.fyear,
a.sex,
a.age,
a.mainspef,
a.speldur,
a.icb,
a.resladst_ons,
a.lsoa11,
a.group,
CASE WHEN a.age BETWEEN 0 AND 4 THEN '0-4'
     WHEN a.age BETWEEN 5 AND 9 THEN '5-9'
     WHEN a.age BETWEEN 10 AND 14 THEN '10-14'
     WHEN a.age BETWEEN 15 AND 19 THEN '15-19'
     WHEN a.age BETWEEN 20 AND 24 THEN '20-24'
     WHEN a.age BETWEEN 25 AND 29 THEN '25-29'
     WHEN a.age BETWEEN 30 AND 34 THEN '30-34'
     WHEN a.age BETWEEN 35 AND 39 THEN '35-39'
     WHEN a.age BETWEEN 40 AND 44 THEN '40-44'
     WHEN a.age BETWEEN 45 AND 49 THEN '45-49'
     WHEN a.age BETWEEN 50 AND 54 THEN '50-54'
     WHEN a.age BETWEEN 55 AND 59 THEN '55-59'
     WHEN a.age BETWEEN 60 AND 64 THEN '60-64'
     WHEN a.age BETWEEN 65 AND 69 THEN '65-69'
     WHEN a.age BETWEEN 70 AND 74 THEN '70-74'
     WHEN a.age BETWEEN 75 AND 79 THEN '75-79'
     WHEN a.age BETWEEN 80 AND 84 THEN '80-84'
     WHEN a.age BETWEEN 85 AND 89 THEN '85-89'
     WHEN a.age >=90 THEN '90+'
     ELSE 'NA'
     END AS age_range,
CASE WHEN a.speldur= 0 THEN '0'
     WHEN a.speldur=1 THEN '1'
     WHEN a.speldur=2 THEN '2'
     WHEN a.speldur=3 THEN '3'
     WHEN a.speldur BETWEEN 4 AND 7 THEN '4-7'
     WHEN a.speldur BETWEEN 8 AND 14 THEN '8-14'
     WHEN a.speldur BETWEEN 15 AND 21 THEN '15-21'
     WHEN a.speldur>=22 THEN '22+'
     ELSE 'NA'
    END AS los_range,
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'alcohol_partially_attributable_acute') 
                 THEN 1 ELSE 0 END as alcohol_partially_attributable_acute,
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'alcohol_partially_attributable_chronic') 
                 THEN 1 ELSE 0 END as alcohol_partially_attributable_chronic, 
 CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'alcohol_wholly_attributable') 
                 THEN 1 ELSE 0 END as alcohol_wholly_attributable,  
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'ambulatory_care_conditions_acute') 
                 THEN 1 ELSE 0 END as ambulatory_care_conditions_acute,
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'ambulatory_care_conditions_chronic') 
                 THEN 1 ELSE 0 END as ambulatory_care_conditions_chronic, 
 CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'ambulatory_care_conditions_vaccine_preventable') 
                 THEN 1 ELSE 0 END as ambulatory_care_conditions_vaccine_preventable,        
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'eol_care_2_days') 
                 THEN 1 ELSE 0 END as eol_care_2_days,
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'eol_care_3_to_14_days') 
                 THEN 1 ELSE 0 END as eol_care_3_to_14_days, 
 CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'falls_related_admissions') 
                 THEN 1 ELSE 0 END as falls_related_admissions,      
 CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'frail_elderly_high') 
                 THEN 1 ELSE 0 END as frail_elderly_high,
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'frail_elderly_intermediate') 
                 THEN 1 ELSE 0 END as frail_elderly_intermediate, 
 CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'intentional_self_harm') 
                 THEN 1 ELSE 0 END as intentional_self_harm,  
 CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'medically_unexplained_related_admissions') 
                 THEN 1 ELSE 0 END as medically_unexplained_related_admissions,
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'medicines_related_admissions_explicit') 
                 THEN 1 ELSE 0 END as medicines_related_admissions_explicit, 
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'medicines_related_admissions_implicit_anti-diabetics') 
                 THEN 1 ELSE 0 END as medicines_related_admissions_implicit_anti_diabetics,      
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'medicines_related_admissions_implicit_benzodiasepines') 
                 THEN 1 ELSE 0 END as medicines_related_admissions_implicit_benzodiasepines,
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'medicines_related_admissions_implicit_diurectics') 
                 THEN 1 ELSE 0 END as medicines_related_admissions_implicit_diurectics, 
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'medicines_related_admissions_implicit_nsaids') 
                 THEN 1 ELSE 0 END as medicines_related_admissions_implicit_nsaids


FROM su_data.nhp.apc a



```

```sql
--- Adding second batch of mitigators

CREATE OR REPLACE TABLE strategyunit.default.SLtemp_apc_mitigator_table2 AS

SELECT *,
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'obesity_related_admissions') 
                 THEN 1 ELSE 0 END as obesity_related_admissions,
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'raid_ae') 
                 THEN 1 ELSE 0 END as raid_ae, 
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'readmission_within_28_days') 
                 THEN 1 ELSE 0 END as readmission_within_28_days, 
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'smoking') 
                 THEN 1 ELSE 0 END as smoking,
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'virtual_wards_activity_avoidance_ari') 
                 THEN 1 ELSE 0 END as virtual_wards_activity_avoidance_ari, 
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'virtual_wards_activity_avoidance_heart_failure') 
                 THEN 1 ELSE 0 END as virtual_wards_activity_avoidance_heart_failure, 
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'zero_los_no_procedure_adult') 
                 THEN 1 ELSE 0 END as zero_los_no_procedure_adult,
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'zero_los_no_procedure_child') 
                 THEN 1 ELSE 0 END as zero_los_no_procedure_child, 
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'emergency_elderly') 
                 THEN 1 ELSE 0 END as emergency_elderly, 
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'raid_ip') 
                 THEN 1 ELSE 0 END as raid_ip, 
CASE WHEN EXISTS (SELECT 1
                 FROM su_data.nhp.apc_mitigators b
                 WHERE a.epikey= b.epikey
                 AND b.strategy= 'stroke_early_supported_discharge') 
                 THEN 1 ELSE 0 END as stroke_early_supported_discharge

FROM strategyunit.default.SLtemp_apc_mitigator_table a
```

```sql
--- Adding in ethnicity information

CREATE OR REPLACE TABLE strategyunit.default.SLtemp_apc_mitigator_table3 AS

SELECT a.*,
b.ethnos

FROM strategyunit.default.SLtemp_apc_mitigator_table2 a

LEFT JOIN hes.silver.apc  b
ON a.epikey = b.epikey AND
a.fyear=b.fyear
```

```sql
-- Adding in IMD information information

CREATE OR REPLACE TABLE strategyunit.default.SLtemp_apc_mitigator_table4 AS

SELECT a.*,
b.imd19_decile

FROM strategyunit.default.SLtemp_apc_mitigator_table3 a

LEFT JOIN su_data.reference.lsoa11_to_imd19  b
ON a.lsoa11 = b.lsoa11 
```

```sql
--- Summarising the episode and beddays for the mitigible activity

CREATE OR REPLACE TABLE strategyunit.default.SL_AF_describing_mitigators_final AS

SELECT 
a.fyear,
a.age_range,
a.sex,
a.ethnos,
a.imd19_decile,
a.los_range,
a.mainspef,
a.icb,
a.resladst_ons,
a.alcohol_partially_attributable_acute,
a.alcohol_partially_attributable_chronic,
a.alcohol_wholly_attributable,
a.ambulatory_care_conditions_acute,
a.ambulatory_care_conditions_chronic,
a.ambulatory_care_conditions_vaccine_preventable,
a.eol_care_2_days,
a.eol_care_3_to_14_days,
a.falls_related_admissions,
a.frail_elderly_high,
a.frail_elderly_intermediate,
a.intentional_self_harm,
a.medically_unexplained_related_admissions,
a.medicines_related_admissions_explicit,
a.medicines_related_admissions_implicit_anti_diabetics,
a.medicines_related_admissions_implicit_benzodiasepines,
a.medicines_related_admissions_implicit_diurectics,
a.medicines_related_admissions_implicit_nsaids,
a.obesity_related_admissions,
a.raid_ae,
a.readmission_within_28_days,
a.smoking,
a.virtual_wards_activity_avoidance_ari,
a.virtual_wards_activity_avoidance_heart_failure,
a.zero_los_no_procedure_adult,
a.zero_los_no_procedure_child,
a.emergency_elderly,
a.raid_ip,
a.stroke_early_supported_discharge,
Count(DISTINCT a.epikey) AS episodes,
Sum(a.speldur) AS beddays

FROM strategyunit.default.SLtemp_apc_mitigator_table4 a

WHERE  a.alcohol_partially_attributable_acute =1 OR
        a.alcohol_partially_attributable_chronic  =1 OR
        a.alcohol_wholly_attributable  =1 OR
        a.ambulatory_care_conditions_acute  =1 OR
        a.ambulatory_care_conditions_chronic  =1 OR
        a.ambulatory_care_conditions_vaccine_preventable  =1 OR
        a.eol_care_2_days  =1 OR
        a.eol_care_3_to_14_days  =1 OR
        a.falls_related_admissions  =1 OR
        a.frail_elderly_high  =1 OR
        a.frail_elderly_intermediate  =1 OR
        a.intentional_self_harm  =1 OR
        a.medically_unexplained_related_admissions  =1 OR
        a.medicines_related_admissions_explicit  =1 OR
        a.medicines_related_admissions_implicit_anti_diabetics  =1 OR
        a.medicines_related_admissions_implicit_benzodiasepines  =1 OR
        a.medicines_related_admissions_implicit_diurectics  =1 OR
        a.medicines_related_admissions_implicit_nsaids  =1 OR
        a.obesity_related_admissions  =1 OR
        a.raid_ae  =1 OR
        a.readmission_within_28_days  =1 OR
        a.smoking  =1 OR
        a.virtual_wards_activity_avoidance_ari  =1 OR
        a.virtual_wards_activity_avoidance_heart_failure  =1 OR
        a.zero_los_no_procedure_adult  =1 OR
        a.zero_los_no_procedure_child  =1 OR
        a.emergency_elderly  =1 OR
        a.raid_ip  =1 OR
        a.stroke_early_supported_discharge =1 OR
        a.fyear IS NOT NULL

GROUP BY
a.fyear,
a.age_range,
a.sex,
a.ethnos,
a.imd19_decile,
a.los_range,
a.mainspef,
a.icb,
a.resladst_ons,
a.alcohol_partially_attributable_acute,
a.alcohol_partially_attributable_chronic,
a.alcohol_wholly_attributable,
a.ambulatory_care_conditions_acute,
a.ambulatory_care_conditions_chronic,
a.ambulatory_care_conditions_vaccine_preventable,
a.eol_care_2_days,
a.eol_care_3_to_14_days,
a.falls_related_admissions,
a.frail_elderly_high,
a.frail_elderly_intermediate,
a.intentional_self_harm,
a.medically_unexplained_related_admissions,
a.medicines_related_admissions_explicit,
a.medicines_related_admissions_implicit_anti_diabetics,
a.medicines_related_admissions_implicit_benzodiasepines,
a.medicines_related_admissions_implicit_diurectics,
a.medicines_related_admissions_implicit_nsaids,
a.obesity_related_admissions,
a.raid_ae,
a.readmission_within_28_days,
a.smoking,
a.virtual_wards_activity_avoidance_ari,
a.virtual_wards_activity_avoidance_heart_failure,
a.zero_los_no_procedure_adult,
a.zero_los_no_procedure_child,
a.emergency_elderly,
a.raid_ip,
a.stroke_early_supported_discharge





```

```sql
CREATE OR REPLACE TABLE strategyunit.default.SL_AF_total_elective_emergency_activity AS

SELECT 
    fyear,
    age_range,
    sex,
    ethnos,
    imd19_decile,
    los_range,
    mainspef,
    icb,
    resladst_ons,
    CASE 
        WHEN (`group`) = "elective" THEN count(distinct epikey) 
        ELSE NULL 
    END AS total_episodes_elective,
    CASE 
        WHEN (`group`) = "elective" THEN sum(speldur) 
        ELSE NULL 
    END AS total_beddays_elective,
        CASE 
        WHEN (`group`) = "non-elective" THEN count(distinct epikey) 
        ELSE NULL 
    END AS total_episodes_emergency,
    CASE 
        WHEN (`group`) = "non-elective" THEN sum(speldur) 
        ELSE NULL 
    END AS total_beddays_emergency

FROM strategyunit.default.SLtemp_apc_mitigator_table4

GROUP BY
    fyear,
    age_range,
    sex,
    ethnos,
    imd19_decile,
    los_range,
    mainspef,
    icb,
    resladst_ons,
    `group`
```
