```sql
--- Table of denominators for providers

CREATE OR REPLACE TABLE strategyunit.default.total_elective_emergency_activity_Dx AS

SELECT 
    fyear,
    provider,
    CASE WHEN age BETWEEN 0 AND 4 THEN '0-4'
     WHEN age BETWEEN 5 AND 9 THEN '5-9'
     WHEN age BETWEEN 10 AND 14 THEN '10-14'
     WHEN age BETWEEN 15 AND 19 THEN '15-19'
     WHEN age BETWEEN 20 AND 24 THEN '20-24'
     WHEN age BETWEEN 25 AND 29 THEN '25-29'
     WHEN age BETWEEN 30 AND 34 THEN '30-34'
     WHEN age BETWEEN 35 AND 39 THEN '35-39'
     WHEN age BETWEEN 40 AND 44 THEN '40-44'
     WHEN age BETWEEN 45 AND 49 THEN '45-49'
     WHEN age BETWEEN 50 AND 54 THEN '50-54'
     WHEN age BETWEEN 55 AND 59 THEN '55-59'
     WHEN age BETWEEN 60 AND 64 THEN '60-64'
     WHEN age BETWEEN 65 AND 69 THEN '65-69'
     WHEN age BETWEEN 70 AND 74 THEN '70-74'
     WHEN age BETWEEN 75 AND 79 THEN '75-79'
     WHEN age BETWEEN 80 AND 84 THEN '80-84'
     WHEN age BETWEEN 85 AND 89 THEN '85-89'
     WHEN age >=90 THEN '90+'
     ELSE 'NA'
     END AS age_range,
    sex,
    CASE 
        WHEN `group` = "elective" THEN count(distinct epikey) 
        ELSE NULL 
    END AS total_episodes_elective,
    CASE 
        WHEN `group` = "elective" THEN sum(speldur) 
        ELSE NULL 
    END AS total_beddays_elective,
        CASE 
        WHEN `group` = "non-elective" THEN count(distinct epikey) 
        ELSE NULL 
    END AS total_episodes_emergency,
    CASE 
        WHEN `group` = "non-elective" THEN sum(speldur) 
        ELSE NULL 
    END AS total_beddays_emergency,
            CASE 
        WHEN (`group` = "non-elective" OR `group` = "elective") THEN count(distinct epikey) 
        ELSE NULL 
    END AS total_episodes,
    CASE 
        WHEN (`group` = "non-elective" OR `group` = "elective")  THEN sum(speldur) 
        ELSE NULL 
    END AS total_beddays


FROM nhp.raw_data.apc

GROUP BY   
    fyear,
    provider,
    sex,
    `group`,
    CASE WHEN age BETWEEN 0 AND 4 THEN '0-4'
     WHEN age BETWEEN 5 AND 9 THEN '5-9'
     WHEN age BETWEEN 10 AND 14 THEN '10-14'
     WHEN age BETWEEN 15 AND 19 THEN '15-19'
     WHEN age BETWEEN 20 AND 24 THEN '20-24'
     WHEN age BETWEEN 25 AND 29 THEN '25-29'
     WHEN age BETWEEN 30 AND 34 THEN '30-34'
     WHEN age BETWEEN 35 AND 39 THEN '35-39'
     WHEN age BETWEEN 40 AND 44 THEN '40-44'
     WHEN age BETWEEN 45 AND 49 THEN '45-49'
     WHEN age BETWEEN 50 AND 54 THEN '50-54'
     WHEN age BETWEEN 55 AND 59 THEN '55-59'
     WHEN age BETWEEN 60 AND 64 THEN '60-64'
     WHEN age BETWEEN 65 AND 69 THEN '65-69'
     WHEN age BETWEEN 70 AND 74 THEN '70-74'
     WHEN age BETWEEN 75 AND 79 THEN '75-79'
     WHEN age BETWEEN 80 AND 84 THEN '80-84'
     WHEN age BETWEEN 85 AND 89 THEN '85-89'
     WHEN age >=90 THEN '90+'
     ELSE 'NA'
     END
```
