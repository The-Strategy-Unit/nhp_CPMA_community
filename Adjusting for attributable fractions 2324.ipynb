```sql

--- Adjust for alcohol chronic
CREATE OR REPLACE TABLE strategyunit.default.SLtemp_apc_mitigator_table63_2324 AS

SELECT *,
CASE WHEN rate_type = 'inverse_alcohol_chronic_rate' THEN 0 ELSE alcohol_partially_attributable_chronic END AS alcohol_partially_attributable_chronic1,
CASE WHEN rate_type = 'inverse_alcohol_chronic_rate' AND rate_value = 0 THEN episodes ELSE episodes * rate_value END AS episodes1,
CASE WHEN rate_type = 'inverse_alcohol_chronic_rate' AND rate_value = 0 THEN beddays ELSE beddays * rate_value END AS beddays1

FROM strategyunit.default.SLtemp_apc_mitigator_table62_2324 a
```
